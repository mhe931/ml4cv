```python
import cv2
import numpy as np
from matplotlib import pyplot as plt

# 1. Loading and converting the image
# OpenCV reads in BGR by default; Matplotlib displays in RGB [cite: 1094-1097, 1104-1105]
img_bgr = cv2.imread('./resources/Example.png', cv2.IMREAD_COLOR)
img_rgb = cv2.cvtColor(img_bgr, cv2.COLOR_BGR2RGB)

# 2. Displaying the image [cite: 1098-1100]
plt.imshow(img_rgb)
plt.axis("off")
plt.show()

```

```python

# 3. Inspecting metadata [cite: 1107-1110]
print(type(img_bgr))    # Output: <class 'numpy.ndarray'> [cite: 10]
print(img_bgr.shape)   # Output: (32, 32, 3) representing (height, width, channels) [cite: 12]
```

```python

# 4. Accessing and modifying specific pixels [cite: 1112, 1116-1117]
print(img_rgb[17, 8])  # Returns [R, G, B] values of the pixel [cite: 1114-1115]
img_rgb[10, 16] = [255, 255, 255] # Changes pixel at row 10, col 16 to white [cite: 1117]
```

```python

# 5. Arithmetic operations (Darkening)
# // is the floor division operator to keep values as integers [cite: 1127, 1133]
img_rgb_dark = img_rgb // 2 
plt.imshow(img_rgb_dark)
plt.show()
```

```python

# 6. Manual Rotation (Main Algorithmic Logic)
# Transposing the image by swapping indices i and j [cite: 1134-1136]
img_rot = np.zeros(shape=(32, 32, 3), dtype=np.uint8) 

for j in range(0, 32):
    for i in range(0, 32):
        for c in range(0, 3):
            # Swapping coordinates (i,j) to (j,i) performs the rotation [cite: 25]
            img_rot[j, i, c] = img_rgb[i, j, c]

plt.imshow(img_rot)
plt.axis("off")
plt.show()
```
